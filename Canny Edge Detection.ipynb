# Canny Edge Detection

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

Read the Lena Image from the current directory as a grayscale image. The second argument tells to read as  grayscale image

```python
img = cv.imread("Lenna.png", 3)
b,g,r = cv.split(img)
img2 = cv.merge([r,g,b])
```

Create the kernel to convolute with the image. This is to filter the image.

```python
kernel = np.ones((5,5), np.float32 )/ 25
print(kernel)
```

```python
filtered_img = cv.filter2D(img2, -1, kernel)
plt.subplot(1,2,1)
plt.imshow(img2)
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2)
plt.imshow(filtered_img)
plt.title('Filtered'), plt.xticks([]), plt.yticks([])
plt.show()
```
