```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("Lenna.png",1)
b,g,r = cv2.split(img)
img2 = cv2.merge([r,g,b])
```

```python
z = np.zeros((512,512), np.uint8)
img3 = cv2.merge([r,g,z])
img4 = cv2.merge([z,g,b])
img5 = cv2.merge([r,z,b])


plt.subplot(1,3,1)
plt.xticks([])
plt.yticks([])
plt.imshow(img3)

plt.subplot(1,3,2)
plt.xticks([])
plt.yticks([])
plt.imshow(img4)

plt.subplot(1,3,3)
plt.xticks([])
plt.yticks([])
plt.imshow(img5)

plt.xticks([])
plt.yticks([])
plt.show()
```
